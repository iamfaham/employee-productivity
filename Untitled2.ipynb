```python
import pandas as pd

# Load the dataset
df = pd.read_csv("Employee_Performance_and_Productivity.csv")

# Drop rows with missing values
df = df.dropna()

# Display the first few rows of the cleaned DataFrame
df.head()
```

```python
df["Hire_Date"] = pd.to_datetime(df["Hire_Date"])
df["Years_Since_Hired"] = 2025 - df["Hire_Date"].dt.year  # Calculate experience

df["Workload_per_Hour"] = df["Projects_Handled"] / df["Work_Hours_Per_Week"]
df["Overtime_Effectiveness"] = df["Performance_Score"] / (df["Overtime_Hours"] + 1)
df["Satisfaction_Level"] = pd.cut(df["Employee_Satisfaction_Score"],
                                  bins=[0, 3, 7, 10],
                                  labels=["Low", "Medium", "High"])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.scatterplot(data=df, x="Workload_per_Hour", y="Performance_Score", hue="Department")
plt.title("Performance vs Workload")
plt.show()
```

```python
resignation_rates = df.groupby("Department")["Resigned"].mean() * 100
resignation_rates.plot(kind="bar", title="Resignation Rate by Department", color="red")
plt.show()
```

```python
sns.boxplot(data=df, x="Remote_Work_Frequency", y="Employee_Satisfaction_Score")
plt.title("Remote Work vs Satisfaction")
plt.show()
```
